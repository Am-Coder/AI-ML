```python
import tensorflow as tf
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
```

```python
def create_dataset():
    df=pd.read_csv('E:/projects/sonar/sonar.csv')
    x=df[df.columns[0:60]].values
    y=df["Class"]
    
    encoder=LabelEncoder()
    hot=OneHotEncoder()
    encoder.fit(y)
#     print(y)
    y = encoder.transform(y)
#     print(y)
    y =hot.fit_transform(y.reshape(-1,1)).toarray()
    print(y)
    print(x.shape)
    return x,y
```

```python
X,Y = create_dataset()
train_x,test_x,train_y,test_y =train_test_split(X,Y,test_size=0.1)
l_rate=0.001
epoch=1000
n_dim=X.shape[1]
n_class=2
nh1=100
nh2=80
nh3=60
nh4=60
x = tf.placeholder(tf.float32,[None,n_dim])
y_ = tf.placeholder(tf.float32,[None,n_class])
b = tf.Variable(initializer=tf.zeros([n_class]))
w = tf.Variable(initializer=tf.random_normal([n_dim, n_class]))
```

```python
def forward_prop(x,weights,bias):
    
    l_1 = tf.add(tf.matmul(x,weights['w1']),bias['b1'])
    l_1 = tf.nn.sigmoid(l_1)
    
    l_2 = tf.add(tf.matmul(l_1,weights['w2']),bias['b2'])
    l_2 = tf.nn.sigmoid(l_2)
    
    l_3 = tf.add(tf.matmul(l_2,weights['w3']),bias['b3'])
    l_3 = tf.nn.sigmoid(l_3)
    
    l_4 = tf.add(tf.matmul(l_3,weights['w4']),bias['b4'])
    l_4 = tf.nn.sigmoid(l_4)
    
    out = tf.add(tf.matmul(l_4,weights['out']),bias['out'])
#     out = tf.nn.sigmoid(out)
    return out
```

```python

weights = {
    'w1':tf.Variable(tf.truncated_normal([n_dim,nh1])),
    'w2':tf.Variable(tf.truncated_normal([nh1,nh2])),
    'w3':tf.Variable(tf.truncated_normal([nh2,nh3])),
    'w4':tf.Variable(tf.truncated_normal([nh3,nh4])),
    'out':tf.Variable(tf.truncated_normal([nh4,n_class])),

}
biases = {
    'b1':tf.Variable(tf.truncated_normal([1,nh1])),
    'b2':tf.Variable(tf.truncated_normal([1,nh2])),
    'b3':tf.Variable(tf.truncated_normal([1,nh3])),
    'b4':tf.Variable(tf.truncated_normal([1,nh4])),
    'out':tf.Variable(tf.truncated_normal([1,n_class]))
}
init = tf.global_variables_initializer()
y=forward_prop(x,weights,biases)
cost_fxn = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=y,labels=y_))
optimizer = tf.train.GradientDescentOptimizer(l_rate).minimize(cost_fxn)
with tf.Session() as sess:
    sess.run(init)
    mse_his=[]
    accu_his=[]
    for i in range(epoch):
        sess.run(optimizer, feed_dict={x: train_x, y_:train_y})
        cost=sess.run(cost_fxn, feed_dict={x: train_x, y_:train_y})
        cos_his=np.append(cos_his,cost)
        correct_prediction=tf.equal(tf.argmax(y,1),tf.argmax(y_,1))
        accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
        pred_y=sess.run(y, feed_dict={x: test_x})
        mse=tf.reduce_mean(tf.square(pred_y-test_y))
        mse_ = sess.run(mse)
        mse_his.append(mse_)
        accuracy = (sess.run(accuracy, feed_dict={x: train_x,y_:train_y}))
        accu_his.append(accuracy)
        print('epoch:',i,'-','cost:',cost,'-MSE',mse_,'Train Accuracy:',accuracy)
```
